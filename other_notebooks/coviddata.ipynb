# COVID Vaccination data with pandas
This example show how to deal with some complications such as dates which are read in a strings instead of datetimes, and numbers that have commas in them!

```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data/Covid-19_Vaccination.csv")
df['Administered_Dose1'].dtype
df
```

```python
import locale
from locale import atof
locale.setlocale(locale.LC_NUMERIC, '')
df.loc[:, "Administered_Dose1"] = df["Administered_Dose1"].map(atof)
df['Date'] = pd.to_datetime(df['Date'],format='%m/%d/%Y')
#df[['Administered_Dose1','Series_Complete_Yes','Booster_Doses_Yes','Booster_Doses_Yes_Last14Days']].applymap(atof)
df
```

```python
df2 = pd.pivot_table(df,values='Administered_Dose1',index=['Date'],columns=['Demographic_category'])
#df2 = pd.pivot_table(df, values=column, index=["date"], columns=["location"])
df2
```

```python
df2[df2.columns[1:14]].plot(figsize=(20,10))
plt.grid()
plt.title("Dose1 by age category",fontsize=24)
```

```python
df2[df2.columns[[23,24,27]]].plot(figsize=(20,10))
plt.grid()
plt.title("Dose 1 by sex",fontsize=24)
```
