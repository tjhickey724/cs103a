```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('names2K.csv')
fdf = df[df['sex']=='F']
pt = pd.pivot_table(fdf,index='year',columns='name',values='count')
pt[['Emily','Hannah','Madison','Olivia','Emma','Ava']].plot(grid=True,figsize=(15,10))
```
